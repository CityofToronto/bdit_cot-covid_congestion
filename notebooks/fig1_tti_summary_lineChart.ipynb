````python
import json

def write_line_json(df, jname):
    o = {
        "tti": {
            "keys": {
                "name": "date",
                "values": list(df.index)
            },
            "tti_actual": [],
            'tti_baseline_avg': [],
            'tti_baseline_05':[],
            'tti_baseline_95':[]
        }
    }
    
    row1=[]
    for val in df['tti_actual']:
        if np.isnan(val):
            row1.append(None)
        else:
            row1.append(val)
#             row1.append(int(val))
            
    row2=[]
    for val in df['tti_baseline_avg']:
        if np.isnan(val):
            row2.append(None)
        else:
            row2.append(val)
#             row2.append(int(val))

    row3=[]
    for val in df['tti_baseline_05']:
        if np.isnan(val):
            row3.append(None)
        else:
            row3.append(val)
    
    row4=[]
    for val in df['tti_baseline_95']:
        if np.isnan(val):
            row4.append(None)
        else:
            row4.append(val)

    o['tti']['tti_actual'] = row1
    o['tti']['tti_baseline_avg'] = row2
    o['tti']['tti_baseline_05'] = row3
    o['tti']['tti_baseline_95'] = row4

    file = open(jname, 'w')
    
    print(o)
    
    json.dump(o, file)
````

````python
jname = '../src/data/fig1_tti_summary.json'
write_line_json(df, jname)
````
